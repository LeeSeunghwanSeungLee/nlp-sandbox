# 1. Pandas Test

## 1) Series

```python
import pandas as pd
```

```python
sr = pd.Series([17000, 18000, 1000, 5000],
               index = ["pizza", "chicken", "coke", "beer"])
```

```python
print("print series")
print('-' * 15)
print(sr)
print("=" * 15)
print(type(sr))
```

```python
print("series value : {}".format(sr.values))
print("=" * 70)
print("series index : {}".format(sr.index))
```

## 2) Data Frame

```python
values = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
index = ['one', 'two', 'three'] # row name
columns = ['A', 'B', 'C'] # field

df = pd.DataFrame(values, index=index, columns=columns)

print('Data Frame :')
print('-'*18)
print(df)
```

```python
print('DataFrame index : {}'.format(df.index))
print('DataFrame Field name: {}'.format(df.columns))
print('DataFrame values :')
print('-'*18)
print(df.values)
print('='*18)
print(type(df.values)) # numpy array type
```

```python
# use list type
data = [
    ['1000', 'Steve', 90.72], 
    ['1001', 'James', 78.09], 
    ['1002', 'Doyeon', 98.43], 
    ['1003', 'Jane', 64.19], 
    ['1004', 'Pilwoong', 81.30],
    ['1005', 'Tony', 99.14],
]

df = pd.DataFrame(data)
print(df)
print(df.index)
print(df.columns)
```

```python
df = pd.DataFrame(data, columns=['student id', 'name', 'score'])
print(df)
```

```python
# use dictionary type
data = {
    'student_id' : ['1000', '1001', '1002', '1003', '1004', '1005'],
    'name' : [ 'Steve', 'James', 'Doyeon', 'Jane', 'Pilwoong', 'Tony'],
    'score': [90.72, 78.09, 98.43, 64.19, 81.30, 99.14]
    }

df = pd.DataFrame(data)
print(df)
```

# 2. Numpy

## 1) np.array() type

```python
import numpy as np
```

```python
# 1 dimension array
vec = np.array([1,2,3,4,5])
print(vec)
```

```python
# 2 dimension array
mat = np.array([[10, 20, 30], [ 60, 70, 80]]) 
print(mat)
print("=" * 15)
print(type(mat))
```

```python
print("vec's dimention :", vec.ndim) # 차원 출력
print("vec's shape : ",vec.shape) # 크기 출력
```

## 2) ndarray initialize

```python
# 2 x 3 matrics compose of zeros
zero_mat = np.zeros((2,3))
print(zero_mat)
```

```python
one_mat = np.ones((2,3))
print(one_mat)
```

```python
same_value_mat = np.full((2,3), 7)
print(same_value_mat)
```

```python
eye_mat = np.eye(3)
print(eye_mat)
```

```python
random_mat = np.random.random((2,2)) # 임의의 값으로 채워진 배열 생성
print(random_mat)
```

## 3) np.arange()

```python
range_vec = np.arange(10)
print(range_vec)
```

```python
n = 2
range_n_step_vec = np.arange(1, 10, n)
print(range_n_step_vec)
```

## 4) reshape()

```python
reshape_mat = np.array(np.arange(30)).reshape((5,6))
print(reshape_mat)
```

## 5) numpy slicing

```python
mat = np.array([[1, 2, 3], [4, 5, 6]])
print(mat)
```

```python
# slice 1st row
slicing_mat = mat[0, :]
print(slicing_mat)
```

```python
# slice 2nd column
slicing_mat = mat[:, 1]
print(slicing_mat)
```

## 6) numpy real number indexing

```python
mat = np.array([[1, 2], [4, 5], [7, 8]])
print(mat)
```

```python
# 1행 0열의 원소
# => 0부터 카운트하므로 두번째 행 첫번째 열의 원소.
print(mat[1, 0])
```

```python
indexing_mat = mat[[2, 1],[0, 1]]
print(indexing_mat)
```

## 7) numpy calculate

```python
x = np.array([1,2,3])
y = np.array([4,5,6])
```

```python
result = x + y
print(result)
# result = np.subtract(x, y)와 동일.
result = x - y
print(result)
# result = np.multiply(result, x)와 동일.
result = result * x
print(result)
# result = np.divide(result, x)와 동일.
result = result / x
print(result)
```

```python
# dot product

mat1 = np.array([[1,2],[3,4]])
mat2 = np.array([[5,6],[7,8]])

mat3 = np.dot(mat1, mat2)
print(mat3)
```

# 3. Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.title('test title')
plt.plot([1,2,3,4],[2,4,8,6])
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show()
```
