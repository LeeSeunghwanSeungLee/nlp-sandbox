```python
en_text = "A Dog RUn Back corner near spare bedrooms"
```

```python
import spacy
spacy_en = spacy.load('en')
```

```python
def tokenize(en_text):
    return [tok.text for tok in spacy_en.tokenizer(en_text)]
```

```python
print(tokenize(en_text))
```
