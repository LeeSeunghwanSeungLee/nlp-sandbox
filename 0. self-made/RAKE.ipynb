## RAKE 를 이용한 text mining

[링크](https://pongdangstory.tistory.com/368)

```python
! pip install python-rake==1.4.4
```

```python
import RAKE
import operator
```

```python
stop_dir = '/content/StopList.txt'
rake_object = RAKE.Rake(stop_dir)

def Sort_Tuple(tup):
    tup.sort(key = lambda x: x[1])
    return tup

subtitles = "blah blah"

keywords = Sort_Tuple(rake_object.run(subtitles))[-10:]
print ("keywords : ", keywords)
```

```python
from rake_nltk import Rake

r = Rake()

r.extract_keywords_from_text(subtitles)

r.get_ranked_phrases()[0 : 10]
```
